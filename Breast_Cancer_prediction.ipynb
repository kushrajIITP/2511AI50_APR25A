```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
import pandas as pd

df = pd.read_csv("hf://datasets/scikit-learn/breast-cancer-wisconsin/breast_cancer.csv")
```

```python
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
df_encoded=pd.get_dummies(df,columns=['diagnosis'],drop_first=True)
df_encoded
```

```python
df_encoded.dtypes
```

```python
X=df_encoded.drop(columns=['diagnosis_M'])
sc=StandardScaler()
X_scale=pd.DataFrame(sc.fit_transform(X),columns=X.columns)
X_scale
```

```python
X_scale.describe()
```

```python
Y=pd.DataFrame(df_encoded.diagnosis_M)
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_scale,Y,test_size=0.2,random_state=4)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
Y_pred = pd.DataFrame(model.predict(X_test))
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
conf_matrix = confusion_matrix(Y_test, Y_pred)
print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")
```
